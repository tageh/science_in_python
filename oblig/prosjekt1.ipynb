<p>Oppgave 5</p>

```python
import numpy as np

listAvstand = []
xkor = []
ykor = []

xstart = 0.0
ystart = 0.0

def avstand(xend,yend,xstart,ystart): 
    return np.sqrt((xend-xstart)**2 + (yend-ystart)**2)

for i in range(4):
    xend = float(input('Skriv en X kordinat: '))
    xkor.append(xend)
    yend = float(input('Skriv en Y kordinat: '))
    ykor.append(yend)
    listAvstand.append(avstand(xend,yend,xstart,ystart))
    xstart = xend
    ystart = yend

def startSlutt(xe,ye,xs,ys):
    xe = xkor[-1]
    ye = ykor[-1]
    xs = xkor[0]
    ys = ykor[0]
    return np.sqrt((xe-xs)**2 + (ye-ys)**2)
    
print('Sum avastander %.4f'% sum(listAvstand))
print('Avstand luftlinje fra startpunkt til sluttpunkt', startSlutt(xkor,ykor,xkor,ykor))
print('X kordinater', xkor, 'Y kordinater', ykor)
print('Avstander:',listAvstand)
```
